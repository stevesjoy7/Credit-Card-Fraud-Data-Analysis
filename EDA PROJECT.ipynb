```python
## Observe the distribution of classes with time
```

```python
# Creating fraudulent dataframe
data_fraud = df[df['Class'] == 1]
# Creating non fraudulent dataframe
data_non_fraud = df[df['Class'] == 0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a distribution plot with updated functions
plt.figure(figsize=(8,5))

# Kernel Density Estimate for fraudulent transactions
sns.kdeplot(data_fraud['Time'], label='fraudulent', fill=False)

# Kernel Density Estimate for non-fraudulent transactions
sns.kdeplot(data_non_fraud['Time'], label='non fraudulent', fill=False)

# Set the labels and title of the plot
plt.xlabel('Seconds elapsed between the transaction and the first transaction')
plt.legend()
plt.show()
```

We do not see any specific pattern for the fraudulent and non-fraudulent transctions with respect to Time. Hence, we can drop the Time column.

```python
# Dropping the Time column
df.drop('Time', axis=1, inplace=True)
```

## Observe the distribution of classes with amount

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a distribution plot with updated functions
plt.figure(figsize=(8, 5))

# Kernel Density Estimate for fraudulent transaction amounts
sns.kdeplot(data_fraud['Amount'], label='fraudulent', fill=False)

# Kernel Density Estimate for non-fraudulent transaction amounts
sns.kdeplot(data_non_fraud['Amount'], label='non fraudulent', fill=False)

# Set the labels and title of the plot
plt.xlabel('Transaction Amount')
plt.legend()
plt.show()
```

We can see that the fraudulent transctions are mostly densed in the lower range of amount, whereas the non-fraudulent transctions are spreaded throughout low to high range of amount.

## Feature Scaling
We need to scale only the Amount column

```python
# Standardization method
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate the Scaler
scaler = StandardScaler()
```

```python
# Import library
from sklearn.model_selection import train_test_split
```

```python
# Putting feature variables into X
X = df.drop(['Class'], axis=1)
```

```python
# Putting target variable to y
y = df['Class']
```

```python
# Splitting data into train and test set 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
# Fit the data into scaler and transform
X_train['Amount'] = scaler.fit_transform(X_train[['Amount']])
X_train.head()
```

## Checking the Skewness

```python
# Listing the columns
cols = X_train.columns
cols
```

```python
# Columns to keep
target_columns = [ 'V1', 'V2', 'V3', 'V4', 'V5', 'Amount', 'Class']

# Create a new DataFrame with the selected columns
cc_data = df[target_columns]
cc_data.head()
```

```python
# Define the columns you want to plot
columns_to_plot = ['V1', 'V2', 'V3', 'V4', 'V5', 'Amount']

# Create subplots for the histograms
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(columns_to_plot):
    row_idx = i // 3
    col_idx = i % 3

    sns.histplot(data=cc_data, x=col, kde=True, color='blue', ax=axes[row_idx, col_idx])
    axes[row_idx, col_idx].set_title(f'Histogram of {col}')
    axes[row_idx, col_idx].set_xlabel(col)
    axes[row_idx, col_idx].set_ylabel('Frequency')

# Remove empty subplots, if any
for i in range(len(columns_to_plot), 2 * 3):
    fig.delaxes(axes.flatten()[i])

plt.show()
```
